```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tqdm.notebook import tqdm

import pickle
```

```python
with open('re_pulled_songs.pkl', 'rb') as f:
    df_details = pickle.load(f)
```

```python
with open('pulled_features_.pkl', 'rb') as f:
    df_features = pickle.load(f)
```

```python
len(df_details)
```

```python
df_old = pd.read_csv('clean_dataset.csv')
df_old.drop(columns='Unnamed: 0', inplace=True)
df_old
```

```python
df_old.info()
```

```python
new_d = {}
new_d['album_name'] = []
new_d['artists'] = []
new_d['available_markets'] = []
new_d['disc_number'] = []
new_d['duration_ms'] = []
new_d['explicit'] = []
new_d['external_ids'] = []
new_d['external_urls'] = []
new_d['href'] = []
new_d['track_id'] = []
new_d['is_local'] = []
new_d['track_name'] = []
new_d['popularity'] = []
new_d['preview_url'] = []
new_d['track_number'] = []
new_d['type'] = []
new_d['uri'] = []
new_d['release_date'] = []
```

```python
df_details[0]['album']['name']
```

```python

for track in df_details:
    for col in track.keys():
        if col == 'album':
            new_d['album_name'].append(track[col]['name'])
            new_d['release_date'].append(track[col]['release_date'])

        elif col == 'id':
            new_d['track_id'].append(track[col])
        elif col == 'artists':
            temp_s = ''
            for c in track[col]:
                temp_s += c['name']
                if len(track[col]) > 1:
                    temp_s += ', '
            new_d['artists'].append(temp_s)
        elif col == 'name':
            new_d['track_name'].append(track[col])

        else:
            new_d[col].append(track[col])
```

```python
df_details_real = pd.DataFrame(new_d).drop(columns=['available_markets','is_local','track_number','uri','type','external_urls','external_ids'])
```

```python
df_details_real
```

```python
d_features = {}
for x in df_features[0].keys():
    d_features[x] = []
```

```python
df_features[0].keys() == d_features.keys()
```

```python
for i in tqdm(range(len(df_features))):
    if df_features[i] is not None:
        for x in d_features.keys():
            if type(df_features[i][x]) == None:
                d_features[x].append(np.nan)
            else:
                d_features[x].append(df_features[i][x])
```

```python
df_features_real = pd.DataFrame(d_features)
```

```python
df_features_real['track_id'] = df_features_real['id']
```

```python
df_features_real.columns
```

```python
df_details_real.columns
```

```python
df_final = pd.merge(df_details_real,df_features_real, on='track_id', how='left')
```

```python
df_final.columns
```

```python
df_final['duration_ms'] = df_final['duration_ms_x']
```

```python
df_final.drop(columns=['duration_ms_x','duration_ms_y','id', 'uri', 'track_href', 'analysis_url', 'type',
                     'disc_number'], inplace=True)
```

```python
df_final.info()
```

```python
final_df_2 = pd.merge(df_final, df_old[['track_id','track_genre']], on='track_id',how='left')
```

```python
final_df_final = final_df_2[['track_id', 'album_name', 'artists', 'track_name', 'release_date',
       'track_genre', 'duration_ms', 'explicit', 'href', 'danceability',
       'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature',
       'preview_url', 'popularity']]
```
