```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df['track_genre'].unique()
```

```python
#this was the final seed getter from the kaggle dataset
final_d = {}
for x in df['track_genre'].unique():
    final_d[x] = []
    
    #checking for top bracket (80-100)
    if df[(df['track_genre'] == x) & (df['popularity'] > 80) & (df['popularity'] <= 100)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 80) &
                             (df['popularity'] <= 100)].sample(1)['track_id'].to_list()[0]))
        
    elif df[(df['track_genre'] == x) & (df['popularity'] > 60) & (df['popularity'] <= 80)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 60) &
                             (df['popularity'] <= 80)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] > 40) & (df['popularity'] <= 60)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 40) &
                             (df['popularity'] <= 60)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] > 20) & (df['popularity'] <= 40)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 20) &
                             (df['popularity'] <= 40)].sample(1)['track_id'].to_list()[0]))
    else:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                         (df['popularity'] > 0) &
                         (df['popularity'] <= 20)].sample(1)['track_id'].to_list()[0])) 
        
    #checking for second bracket (60-80)    
    if df[(df['track_genre'] == x) & (df['popularity'] > 60) & (df['popularity'] <= 80)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 60) &
                             (df['popularity'] <= 80)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] > 40) & (df['popularity'] <= 60)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 40) &
                             (df['popularity'] <= 60)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] > 20) & (df['popularity'] <= 40)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 20) &
                             (df['popularity'] <= 40)].sample(1)['track_id'].to_list()[0]))
    else:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                         (df['popularity'] > 0) &
                         (df['popularity'] <= 20)].sample(1)['track_id'].to_list()[0])) 
    
    
    #checking for third bracket (40-60)    
    if df[(df['track_genre'] == x) & (df['popularity'] > 40) & (df['popularity'] <= 60)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 40) &
                             (df['popularity'] <= 60)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] > 20) & (df['popularity'] <= 40)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 20) &
                             (df['popularity'] <= 40)].sample(1)['track_id'].to_list()[0]))
    else:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                         (df['popularity'] > 0) &
                         (df['popularity'] <= 20)].sample(1)['track_id'].to_list()[0])) 
        
    #checking for fourth bracket (20-40)    
    if df[(df['track_genre'] == x) & (df['popularity'] > 20) & (df['popularity'] <= 40)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 20) &
                             (df['popularity'] <= 40)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] >= 0) & (df['popularity'] <= 20)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] >= 0) & 
                             (df['popularity'] <= 20)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] > 40) & (df['popularity'] <= 60)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 40) &
                             (df['popularity'] <= 60)].sample(1)['track_id'].to_list()[0]))
    
    #getting last bracket  
    if df[(df['track_genre'] == x) & (df['popularity'] >= 0) & (df['popularity'] <= 20)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] >= 0) & 
                             (df['popularity'] <= 20)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] > 20) & (df['popularity'] <= 40)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 20) &
                             (df['popularity'] <= 40)].sample(1)['track_id'].to_list()[0]))
    elif df[(df['track_genre'] == x) & (df['popularity'] > 40) & (df['popularity'] <= 60)].shape[0] > 0:
        final_d[x].append(str(df[(df['track_genre'] == x) &
                             (df['popularity'] > 40) &
                             (df['popularity'] <= 60)].sample(1)['track_id'].to_list()[0]))
    
```
