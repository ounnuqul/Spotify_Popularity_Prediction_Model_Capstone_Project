```python
#Load the data
df = pd.read_csv('dataset.csv')
df.drop(columns='Unnamed: 0', inplace=True)
```

```python

df1 = pd.DataFrame(df.groupby('track_id')['track_genre'].apply(list)).reset_index()
```

```python
for i in range(df1.shape[0]):
    df1['num_genres'][i] = len(df1['track_genre'][i])
```

```python
df1['genres_str'] = ''
for i in range(df1.shape[0]):
    s = ''
    for x in df1['track_genre'][i]:
        
        s += x
        s += ' '
        
    df1['genres_str'][i] = s
```

```python
df['track_genre'] = df['track_id'].map(df1.set_index('track_id')['genres_str'])
df['num_genres'] = df['track_id'].map(df1.set_index('track_id')['num_genres'])
```
